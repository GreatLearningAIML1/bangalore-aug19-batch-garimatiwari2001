### Find out the no of male & female employees in the company

*Hint: Use can use a condition to find out the no of 1s and 0s in gender and sum it*

```python
# 1 for female 2 for male
femaleCount = 0 
maleCount = 0

for x in HrExternal['Gender']:
    if x==1:
        femaleCount=femaleCount+1
    elif x==2:
        maleCount=maleCount+1

print("Female Employees= ",femaleCount,"\nMale Employees= ",maleCount)
```

### Find out the IQR for the column - Age 

*Hint: IQR - Interquartile range = 75th percentile - 25th percentile* 

```python
AgeCol=HrExternal.sort_values('Age')['Age']
q1, q3= np.percentile(AgeCol,[25,75])
print('Q1: ',q1)
print('Q3: ',q3)
iqr = q3-q1
print('IQR: ',iqr)
```

### Find out the mean, median and variance for the column - DailyRate 

```python
DailyRate = HrExternal.sort_values('DailyRate')['DailyRate']
print("Mean: ",DailyRate.mean())
print("Median: ",DailyRate.median())
print("Variance: ",DailyRate.var())
```
